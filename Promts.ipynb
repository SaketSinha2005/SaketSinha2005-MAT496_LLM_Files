```python
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv
load_dotenv()
```

```python
model=init_chat_model('llama-3.1-8b-instant',model_provider='groq')
```

```python
prompt_template = """Translate the given text in to the given language: 
Text: {text}
Lang: {lang}
"""
```

```python
prompt = prompt_template.format(text = "hey how are you doing?", lang="hindi")
print(prompt)
```

```python
response =  model.invoke(prompt)
response
```

```python
context = '''Alice is a final-year computer science student who has always been fascinated by the intersection of technology and sustainability. For her capstone project, she is exploring how renewable energy—particularly solar power—can be integrated into smart cities to reduce carbon emissions and improve efficiency.

During her research, Alice found that while solar panels can generate significant amounts of electricity during the day, energy demand in cities often peaks in the evening and at night, when solar production drops. This mismatch leads to challenges in balancing supply and demand.

She also learned about battery storage systems, which can store excess solar energy produced during the day and release it later when demand is higher. However, these systems face limitations such as high costs, limited lifespans, and environmental concerns over battery materials.

To address these issues, Alice is investigating hybrid solutions, where solar energy is paired with other renewables (like wind) and supported by AI-driven energy management systems that can forecast demand, optimize storage, and minimize waste. She is also curious about how governments and city planners can incentivize the adoption of such systems, ensuring that renewable energy becomes both practical and affordable for urban populations.'''

prompt_template = """
Context:
{context}

Question:
{question}

Answer:
"""

prom = prompt_template.format(
    context=context,
    question="What are the most effective ways to combine solar energy, battery storage, and AI-driven energy management to make smart cities more sustainable and reliable?"
)

response = model.invoke(prom)
print(response.content if hasattr(response, "content") else response)
```

```python
with open("article.txt", 'r') as file:
    new_context_from_file = file.read()
```

```python
prompt_template = """You are an assistant for finding words starting with specific characters.
Question: {question} 
Context: {context} 
Answer:"""

prompt = prompt_template.format(
    context = new_context_from_file, 
    question="Give me all the words from this article that starts with a "
)

response = model.invoke(prompt)
print(response.content)
```
